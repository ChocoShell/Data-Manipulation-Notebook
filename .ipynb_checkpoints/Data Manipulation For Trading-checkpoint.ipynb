## Numpy Arrays

#### Empty Array

```python
print np.empty((5, 4, 3)) # 5 rows 4 cols 3 depth
```

#### Array full of 1s

```python
print np.ones((5, 4), dtype=np.int_) # set datatype to uinteger. Can do the same thing with np.zeros()
```

#### Generate an array full of random floats from 0.0 to 1.0

```python
print np.random.random((5,4)) # Prefered
# np.random.rand(5, 4)
```

#### Sample numbers from a Gaussian (normal) distribution

```python
print np.random.normal(size=(2,3)) # standard normal (mean=0, std=1)
print np.random.normal(50, 10, size=(2,3)) # standard normal (mean=50, std=10)
```

#### Random Integer arrays

```python
print np.random.randint(10) # 1 integer from 0 to 10
print np.random.randint(-10, 1) # same as above [low, high)
print np.random.randint(0, 10, size=5) # 5 numbers in 1d array
print np.random.randint(0, 10, size=(2,3)) # 2x3 array of random ints
```

#### Playing with Shape

```python
a = np.random.random((5,4))
print a.shape # gives the shape (5,4) row col
```

```python
print a.shape[0] # num of rows
print a.shape[1] # num of cols
```

```python
print len(a.shape) # Dimension of array (2)
print a.size # total number of elements 5*4 = 20
print a.dtype # type of data in array float64
```

#### Random Seed

```python
np.random.seed(693) # seed random number generator to get same random numbers each time
a = np.random.randint(0, 10, size=(5,4)) # Random numbers from 0 to 10.
print a
```

```python
print a.sum() # Sum of all elements
print a.sum(axis=0) # sum of each column
print a.sum(axis=1) # sum of each row
```

```python
print a.min(axis=0) # min of each column
print a.max(axis=1) # max of each row
print a.mean() # average of all elements in column
```

#### Slicing

```python
print a[:, 0:3:2] # Get all rows and columns 0,1,2 but only in steps of 2 so 0,2
```

```python
a[0,:] = 2 # Replace entire row with 2's
print a
```

```python
a[:, 3] = [1,2,3,4,5] # Dimension must be the same
print a
```

#### Masking

```python
a = np.random.randint(10, size=(5,2))
mean = a.mean()
print a
print mean
```

```python
# masking
print a[a<mean]
```

#### Backfill and Front Fill Missing Data

```python
symbols = ["SPY", "AAPL", "AMZN"]
df = get_data(symbols, dates, normal=True)
plot_data(df)
```

```python
df.fillna(method='ffill', inplace=True)

df.fillna(method='bfill', inplace=True)
plot_data(df)
```

```python
def compute_daily_returns(df):
    """Compute and return the daily return values."""
    daily_returns = df.copy()
    daily_returns[1:] = (df[1:]/df[:-1].values) - 1
    daily_returns.iloc[0,:] = 0 # Set daily returns for row 0 to 0
    return daily_returns

daily_returns = compute_daily_returns(df)
plot_data(daily_returns)
```

```python
# Plot a histogram
daily_returns.hist(bins=20) # 20 bins
mean = daily_returns['SPY'].mean()
std = daily_returns['SPY'].std()

plt.axvline(mean)
plt.axvline(std)
plt.axvline(-std)
plt.show()
```

```python
print daily_returns['SPY'].kurtosis() # How thick the ends are.  Higher STD = higher volatility.
```

```python
daily_returns.plot(kind='scatter', x='GOOG', y='FB')
plt.show()
```
